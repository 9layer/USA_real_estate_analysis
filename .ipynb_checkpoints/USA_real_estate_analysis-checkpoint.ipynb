```python
# df['prev_sold_date'] = pd.to_datetime(df['prev_sold_date'])
# После попытки конвертации всплывает ошибка date out of bounds: '3019-04-02'
# скорей всего имеется ввиду 2019-04-02

# Ищем строку с ошибкой
df[(df['prev_sold_date']=='3019-04-02')]
```

```python
# Ищем индекс строки с ошибкой
bad_date_index = df[(df['prev_sold_date']=='3019-04-02')].index
bad_date_index
# переписываем строку используя .loc
df.loc[bad_date_index, 'prev_sold_date'] = '2019-04-02'
```

### Удаление строк с нулевыми значениями 

```python

df = df.dropna(subset=['brokered_by', 'price', 'street', 'city', 'state', 'zip_code']).copy()
```

```python
len(df)
```

```python
df.isna().mean()*100
```

### Добавление новых столбцов

```python
# Количество квадратных метров в доме
df['house_size_m2'] = (df["house_size"] * 0.092903).round(2)
```

```python
# Цена 1 фута (реалистичные значения от 50$ до 150$)
df["price_per_sqft"] = df["price"] / df["house_size"]
```

```python
full_df = df
```

### Поиск выбросов

```python
# Удаляем аномальные строки, где квадратура дома меньше 100 футов (~9.3м) и все NaN
# Теряем большое количество строк, потому что очень много NaN в колонке 
df_known_size = df[(df["house_size"].notna()) & (df["house_size"] >= 100)]
df_unknown_size = df[df["house_size"].isna()]                                                 
```

```python
len(df_known_size)+len(df_unknown_size)
```

```python
df = df_known_size[df_known_size['price_per_sqft'] >= 20]
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df[df["price_per_sqft"].between(20, 2000)]["price_per_sqft"], bins=50, kde=True, color='skyblue')
plt.xlabel("Цена за f² ($/f²)")
plt.ylabel("Количество объектов")
plt.title("Распределение цены за квадратный метр")
plt.show()
```

```python
df = df[(df["price"] >= 10_000) & (df["price"] <= 500_000_000)]
```

### Подключаем SQLite3

```python
con = sqlite3.connect('db')
cur = con.cursor()
df.to_sql('real_estate_with_known_size',con,index=False,if_exists='replace')
full_df.to_sql('real_estate_full',con,index=False,if_exists='replace')
```

```python
def select(sql):
    return pd.read_sql(sql, con)
```

```python
data_dir = os.path.join(os.getcwd(), 'data')
os.makedirs(data_dir, exist_ok=True)
```

```python
file_path = os.path.join(data_dir, 'df.csv')
df.to_csv(file_path, index=False)
```

### Средняя, минимальная и максимальная цена по штату

```python
sql = '''
SELECT state, ROUND(AVG(price)) AS avg_price,
MIN(price) AS min_price,
printf("%.f", MAX(price)) AS max_price
FROM real_estate_with_known_size
GROUP BY state
ORDER BY avg_price DESC
'''
```

```python
df3 = select(sql)
select(sql)
```

```python
file_path = os.path.join(data_dir, 'df3.csv')
```

### Медианная цена по штатам

```python
sql = '''
SELECT state, AVG(price) AS median_price
    FROM
(SELECT state, price,
           ROW_NUMBER() OVER (PARTITION BY state ORDER BY price) AS row_num,
           COUNT(*) OVER (PARTITION BY state) AS total_count
        FROM real_estate_with_known_size) AS sub
    WHERE row_num IN ((total_count + 1) / 2, (total_count + 2) / 2)
    GROUP BY state;
'''
```

```python
df4 = select(sql)
select(sql)
```

```python
file_path = os.path.join(data_dir, 'df4.csv')
df4.to_csv(file_path, index=False)
```

### Средняя цена за квадратный фут по штатам

```python
sql = '''
SELECT state, ROUND(AVG(price_per_sqft)) AS avg_price_per_sqft 
FROM real_estate_with_known_size
GROUP BY state
ORDER BY state
'''
```

```python
df5 = select(sql)
select(sql)
```

```python
file_path = os.path.join(data_dir, 'df5.csv')
df5.to_csv(file_path, index=False)
```

### Топ 10 самых дорогих городов по средней цене дома

```python
sql = '''
WITH valid_cities AS
(SELECT city
FROM real_estate_with_known_size
GROUP BY city
HAVING COUNT(*) > 100)

SELECT city, ROUND(AVG(price)) avg_price
FROM real_estate_with_known_size
WHERE city IN valid_cities
GROUP BY city
ORDER BY avg_price DESC
LIMIT 100
'''
```

```python
df6 = select(sql)
select(sql)
```

```python
file_path = os.path.join(data_dir, 'df6.csv')
df6.to_csv(file_path, index=False)
```
